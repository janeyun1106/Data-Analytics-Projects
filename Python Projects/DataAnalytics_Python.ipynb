### <font color = 'blue'>B. First Merges - Combine Accounts with the Bank and Ledger Data</font>

##### <font color = 'blue'>
As you investigate the data, you see that you will need to combine the Accounts data with both the Bank and Ledger data respectively, so you have common fields to use when you investigate. The goal is to get the Accounts fields into both the Bank and Ledger Data so you will then be able to compare the Ledger Data to the Bank Data to confirm that the money was transferred to the correct account.
    
*Hint 1*: For combining the `AccountData` and `BankData`, you will need to combine on two columns.
    
*Hint 2*: For combining the `AccountData` and `LedgerData`, you will need to extract the `Entity ID` from the `Invoice ID` to be able to merge the two data sets.
</font>

````python
#Merge Account and ledger data
count_of_row=LedgerData.shape[0]
entity_list=[]
i=0
#apply split function to every row in LedgerData
for i in range(count_of_row):
    L_entity=LedgerData['Invoice ID'].iloc[i].split('-')[0]
    entity_list.append(L_entity)
    i=i+1

LedgerData['Entity ID']=entity_list
print(len(entity_list))
LedgerData['Entity ID'] = pd.to_numeric(LedgerData['Entity ID'], errors='coerce').astype(int)
print(LedgerData.dtypes)
print(LedgerData.shape)
#Merge LedgerData and Accounts
mergedData_2=Accounts.merge(LedgerData,'inner',left_on='Entity ID',right_on='Entity ID')
print(mergedData_2.shape)
mergedData_2.head()
````

````python
#your code here; merge Accounts and Bank Data
mergedData_1=Accounts.merge(BankData,'left',left_on=['Bank Account','Routing Number'],right_on=['Bank Account','Routing Number'])
print(mergedData_1.shape)
mergedData_1.head()
````
